```python
from skimage import io
photo = io.imread('/home/afroz/Pictures/book4.jpeg')  # Reading it as a numpy array
type(photo)  # We take the photo as an array
```

```python
photo.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
plt.imshow(photo[::-1])  # Start, Stop and step , here the step is -1(We have reversed the steps in this image)
```

```python
plt.imshow(photo[: , ::-1])  # We have reversed the columns
```

```python
plt.imshow(photo[350:500, 100:200 ])  # It gives a cross-section of the image
```

```python
plt.imshow(photo[::2, ::2])  # For every other row and every other column we halved the size of image
```

```python
photo 
photo_sin = np.sin(photo)  # We take the sin of every single element in your photo
photo_sin
```

```python
print(np.sum(photo))  # We can get the sum of the photo array elements
print(np.prod(photo))  
print(np.mean(photo))  # Mean value of the array elements in 'photo'
print(np.std(photo))  # Standard deviation
print(np.var(photo))  # Variance
print(np.min(photo))  # Minimum value
print(np.max(photo))  # Maximum value
print(np.argmin(photo))  # Index value of the minimum value
print(np.argmax(photo))  # Index value of the maximum value
```
